```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# URL of page to be scraped
url = 'https://www.thebalance.com/the-average-credit-score-by-state-4161310'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Explore returned html
print(soup.prettify())
```

```python
# pull the table of interest
table = soup.find("table",class_="mntl-sc-block-table__table")
```

```python
# look at the table structure
print(table.prettify())
```

```python
# pull the table rows
rows = table.find_all("tr")
```

```python
rows[0]
```

```python
# test loop method with first row
tds = rows[0].find_all("td")
```

```python
# template for larger for loop
test_list = []
for td in tds:
    test_list.append( td.text )
```

```python
# one row works
test_list
```

```python
# apply logic to all the rows, extract table data to list of lists
state_credit_info = []
for row in rows:
    state_info = []
    tds = row.find_all("td")
    for td in tds:
        state_info.append( td.text )
    state_credit_info.append( state_info )
    
```

```python
# check list of lists
state_credit_info
```

```python
# use lists of lists to create dataframe
state_credit_df = pd.DataFrame( state_credit_info )
```

```python
# check data frame
state_credit_df
```

```python
# export to csv for transform step
state_credit_df.to_csv("../DataFiles/stateCredit.csv")
```
