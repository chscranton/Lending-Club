```python
#Dependencies
import pandas as pd
from datetime import datetime
import gzip
```

```python
#Read csv file and store in to dataframe
df = pd.read_csv("loan.csv")
df
```

```python
#Pulling out desired columns
new_df = df[["loan_amnt", "term", "int_rate", "issue_d", "purpose", "zip_code", "addr_state", "annual_inc", "home_ownership"]]
new_df
```

```python
#Removing 'months' from each row in the "Term" column
new_df["term"] = new_df["term"].str.replace(" months", "")
```

```python
#Function to convert dates
def format_date(el):
    return datetime.strptime(el, '%b-%Y').strftime('%m-%Y')


# apply conversion to desired column
new_df["issue_d"] = new_df["issue_d"].apply(format_date)
```

```python
#Rename columns
df_final = new_df.rename(columns={
    "issue_d":"issue_date",
    "addr_state":"state"
})
```

```python
#Export to compressed csv
df_final.to_csv('clean_data.csv.gz', index=True, compression='gzip')
```
